```python
Map = geemap.Map()
countries = ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017")
roi = countries.filter(ee.Filter.eq("countr_na", "Swaziland"))
Map.addLayer(roi, {}, "Swaziland")
Map.centerObject(roi, 6)
Map
```

```python
dataset = ee.ImageCollection("CSIC/SPEI/2_9").filterDate('2022-12-01', '2023-01-01')

visualization = {
    'min': -2.33,
    'max': 2.33,
}

roi = countries.filter(ee.Filter.eq("countr_na", "Singapore"))
Map = geemap.Map()
Map.addLayer(dataset, visualization, 'True Color (321)')
Map.centerObject(roi, 8)
Map
```
